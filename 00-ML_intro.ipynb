Data is collected in a training set defined as:

$\mathcal{T} = \{(x_i, y_i)\}_{i=1}^n$

dove n è il numero di campioni osservati. Ogni coppia rappresenta una realizzazione della relazione tra input e output.
Ipotesi: i campioni in T sono indipendenti e identicamente distribuiti.

# Variabili di Input

Gli input sono rappresentati come vettori:

$x = [x_1, x_2, \ldots, x_p] \in \mathbb{R}^p$

dove ogni componente xj corrisponde a una **feature** che descrive un aspetto dell'osservazione. La dimensione p è il numero di feature.  

Se la variabile di input assume valori ordinati e significativi, è una variabile **numerica**. Può essere continua o discreta e le operazioni aritmetiche e le distanze sono ben definite.

Se la variabile di input assume un insieme finito di etichette come valori senza un ordine intrinseco, è una variabile **categorica**. Le etichette sono *identificatori* piuttosto che quantità.

# Variabili di Output
La variabile di output y è generalmente uno scalare.
I problemi di supervised learning sono comunemente categorizzati secondo la natura della variabile di output.

Se l'output è numerico abbiamo a che fare con problemi di **regression**.  L'obiettivo della regressione è imparare una funzione che mappa gli input a output con valori reali.

$y ∈ R$


Se l'output assume un insieme finito di classi abbiamo a che fare con problemi di **classification**.

$y ∈ {1, . . . ,K}$

L'obiettivo è assegnare ogni input a una delle classi disponibili.  
Sia la *classification* che la *regression* appartengono al paradigma di machine learning del **supervised learning**.

# Supervised Learning Workflow
# Workflow di Supervised Learning
A supervised learning workflow typically consists of:  
Un workflow di supervised learning consiste tipicamente di:  
• Raccogliere dati di addestramento etichettati $(xi, yi)$.  
• Scegliere un algoritmo di apprendimento e un tipo di modello.  
• Addestrare il modello $\hat{f}$ adattando i suoi parametri per minimizzare l'errore di predizione.  
• Utilizzare il modello addestrato per fare predizioni su nuovi input x⋆.  
## Prediction
## Predizione
At prediction time, the learning scenario is the following:
Al momento della predizione, lo scenario di apprendimento è il seguente:
• a new input $x^∗$ is observed;  
• viene osservato un nuovo input $x^∗$;  
• l'output vero corrispondente $y^∗$ è sconosciuto;  
• il modello produce una predizione  
$ \hat{y}^∗ =  f(x^∗) $


## Generalization and Overfitting
## Generalizzazione e Overfitting
To be useful, models must not just fit the training data but generalize to new, unseen data.  
Per essere utili, i modelli devono non solo adattarsi ai dati di addestramento ma anche generalizzare a dati nuovi e non visti.  
L'overfitting si verifica quando il modello cattura il rumore o le peculiarità del set di addestramento piuttosto che il modello sottostante, risultando in prestazioni scarse su nuovi input.  
Per monitorare e prevenire l'overfitting, le tecniche includono:  
• Cross-validation.  
• Regularization methods.  
• Early stopping during training.  
• Early stopping durante l'addestramento.  
Although regression and classification differ in the type of output and in the evaluation metrics used, many learning principles, such as generalisation, model complexity, and overfitting, apply to both settings.
Alacune regression che classification differiscono nel tipo di output e nelle metriche di valutazione utilizzate, molti principi di apprendimento, come generalisation, model complexity e overfitting, si applicano a entrambe le impostazioni.
